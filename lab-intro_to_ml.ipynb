```python
# importar librerías
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Comprobar la dimensión del dataset**

```python
# Comprobar la dimensión del dataset
spaceship.shape
```

**Comprobar los tipos de datos**

```python
# Comprobar los tipos de datos
spaceship.dtypes
```

**Comprobar valores nulos**

```python
# Comprobar valores nulos por columna
spaceship.isna().sum()
```

Existen varias estrategias para tratar los valores nulos:

- Eliminar filas o columnas que contengan valores nulos.
- Rellenar los valores nulos con un valor concreto, como la media en variables continuas o la moda en variables categóricas.
- Rellenar los valores nulos mediante un algoritmo.

Para este ejercicio, como la cantidad de valores nulos es baja, eliminaremos las filas que contengan cualquier valor nulo.

```python
# Eliminar filas que contienen valores nulos
spaceship_clean = spaceship.dropna()

spaceship_clean.shape
```

**KNN**

K-Nearest Neighbors es un algoritmo basado en distancias, por lo que requiere que todas las **variables de entrada sean numéricas**.

Por eso, seleccionaremos solo las columnas numéricas como variables predictoras.

```python
# Seleccionar solo columnas numéricas como variables predictoras
# KNN se basa en distancias y necesita variables de entrada numéricas
X = spaceship_clean.select_dtypes(include=np.number)

X.head()
```

Ahora definimos también nuestra variable objetivo.

```python
# Definir la variable objetivo
# Transported es categórica/binaria, por lo que es un problema de clasificación
y = spaceship_clean["Transported"]

y.value_counts()
```

**Train Test Split**

Ahora que hemos separado los datos en **variables predictoras** y **variable objetivo**, y que hemos importado la función **train_test_split**, dividimos `X` e `y` en `X_train`, `X_test`, `y_train` e `y_test`. El 80% de los datos irá al conjunto de entrenamiento y el 20% al conjunto de test.

```python
# Dividir X e y en conjuntos de entrenamiento y test
# 80% entrenamiento y 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=0
)

print("Dimensión de X_train:", X_train.shape)
print("Dimensión de X_test:", X_test.shape)
print("Dimensión de y_train:", y_train.shape)
print("Dimensión de y_test:", y_test.shape)
```

**Selección del modelo**

En este ejercicio utilizaremos **KNN** como modelo predictivo.

Debemos elegir entre **clasificador** o **regresor**. Para decidirlo, hay que tener en cuenta el tipo de variable objetivo.

Inicializamos una instancia de KNN sin fijar ningún hiperparámetro.

```python
# Inicializar un clasificador KNN
# Usamos clasificador porque la variable objetivo es categórica/binaria
knn = KNeighborsClassifier()
knn
```

Entrenamos el modelo con nuestros datos.

```python
# Entrenar el modelo
knn.fit(X_train, y_train)
```

Evaluamos el modelo.

```python
# Evaluar el modelo
train_score = knn.score(X_train, y_train)
test_score = knn.score(X_test, y_test)

print("Train score:", train_score)
print("Test score:", test_score)
```

```python
# Probar diferentes valores de k
for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    score = knn.score(X_test, y_test)
    print("k:", k, "Test score:", score)
```

### Interpretación

La variable objetivo `Transported` es binaria (`True` / `False`), por lo que se trata de un problema de **clasificación**.

Por ese motivo usamos `KNeighborsClassifier`.

El modelo se evalúa con `.score(X_test, y_test)`, que devuelve el **accuracy**: la proporción de predicciones correctas sobre el conjunto de test.

La separación train/test es importante porque permite entrenar el modelo con una parte de los datos y evaluarlo con datos no vistos por el modelo.
