# LAB | Introducción a Machine Learning

En este lab vamos a construir un primer modelo de Machine Learning usando **K-Nearest Neighbors (KNN)** y la separación **train/test split**.

El objetivo es entrenar un modelo de clasificación y obtener una métrica que nos permita hacernos una idea de su rendimiento.


## 1. Cargar los datos

Trabajaremos con el dataset **Spaceship Titanic**.

Datos:
https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv

Metadata:
https://github.com/data-bootcamp-v4/data/blob/main/spaceship_titanic.md


```python
# Importar librerías necesarias para el lab
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Cargar el dataset
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")

# Mostrar las primeras filas para comprobar que la carga ha sido correcta
spaceship.head()
```

```python
# Comprobación rápida de la carga del dataset
print("Filas y columnas:", spaceship.shape)
```

## 2. Comprobar la dimensión del dataset

Usamos `.shape` para ver cuántas filas y columnas tiene el DataFrame.


```python
# Comprobar la dimensión del dataset
spaceship.shape
```

## 3. Comprobar los tipos de datos

Antes de preparar el modelo, revisamos los tipos de datos para identificar qué columnas son numéricas y cuáles son categóricas.


```python
# Comprobar los tipos de datos
spaceship.dtypes
```

```python
# Vista general del DataFrame: columnas, tipos de datos y valores no nulos
spaceship.info()
```

## 4. Comprobar valores nulos

Revisamos los valores nulos porque los modelos de Machine Learning no pueden trabajar directamente con datos faltantes.


```python
# Comprobar valores nulos por columna
spaceship.isna().sum()
```

Existen varias estrategias para tratar valores nulos:

- Eliminar filas o columnas con valores nulos.
- Rellenar los valores nulos con una medida como la media, la mediana o la moda.
- Usar métodos más avanzados de imputación.

Para este ejercicio, como la cantidad de valores nulos es baja, eliminaremos las filas que contengan cualquier valor nulo.


```python
# Eliminar filas con valores nulos
spaceship_clean = spaceship.dropna()

# Comprobar la nueva dimensión después de eliminar nulos
print("Dimensión original:", spaceship.shape)
print("Dimensión después de dropna:", spaceship_clean.shape)
```

## 5. KNN

K-Nearest Neighbors es un algoritmo basado en distancias. Por ese motivo, las variables de entrada deben ser numéricas.

En este lab seleccionaremos solamente las columnas numéricas como variables predictoras.


```python
# Seleccionar solo columnas numéricas como variables predictoras
X = spaceship_clean.select_dtypes(include=np.number)

# Revisar las primeras filas de X
X.head()
```

```python
# Comprobar qué columnas numéricas estamos usando como predictores
X.columns
```

Seleccionamos solo variables numéricas porque KNN calcula distancias entre observaciones.

Antes de entrenar, revisamos las columnas incluidas en `X` para asegurarnos de qué variables predictoras está usando el modelo.


Ahora definimos la variable objetivo.

La columna `Transported` indica si el pasajero fue transportado o no. Como es una variable binaria (`True` / `False`), se trata de un problema de **clasificación**.


```python
# Definir la variable objetivo
y = spaceship_clean["Transported"]

# Comprobar la distribución de la variable objetivo
y.value_counts()
```

## 6. Train Test Split

Dividimos los datos en conjunto de entrenamiento y conjunto de test.

- El conjunto de entrenamiento se usa para entrenar el modelo.
- El conjunto de test se usa para evaluar el modelo con datos que no ha visto.

En este lab usamos una partición 80/20.


```python
# Dividir X e y en conjunto de entrenamiento y conjunto de test
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=0
)

# Comprobar las dimensiones resultantes
print("Dimensión de X_train:", X_train.shape)
print("Dimensión de X_test:", X_test.shape)
print("Dimensión de y_train:", y_train.shape)
print("Dimensión de y_test:", y_test.shape)
```

## 7. Selección del modelo

En este ejercicio utilizaremos **KNN** como modelo predictivo.

Debemos elegir entre clasificador o regresor. Como la variable objetivo `Transported` es binaria, usamos un **clasificador**.


```python
# Inicializar un clasificador KNN
# Usamos clasificador porque la variable objetivo es categórica/binaria
knn = KNeighborsClassifier()
```

En esta celda inicializamos el modelo KNN. La salida del modelo solo mostraría sus parámetros por defecto, pero eso todavía no indica si funciona bien o mal.

Para evaluarlo correctamente, después debemos entrenarlo con los datos de entrenamiento y calcular métricas de rendimiento sobre el conjunto de test.


## 8. Entrenar el modelo

Entrenamos el modelo con los datos de entrenamiento.


```python
# Entrenar el modelo
knn.fit(X_train, y_train);
```

## 9. Evaluar el modelo

Usamos `.score()` para obtener el **accuracy**.

En un modelo de clasificación, el accuracy representa la proporción de predicciones correctas.


```python
# Evaluar el modelo en train y test
train_score = knn.score(X_train, y_train)
test_score = knn.score(X_test, y_test)

print("Train score:", train_score)
print("Test score:", test_score)
```

### Interpretación de la evaluación

La métrica principal usada aquí es el **accuracy**, que indica la proporción de predicciones correctas.

El resultado en test es especialmente importante porque mide cómo funciona el modelo con datos que no ha visto durante el entrenamiento.

Comparar el score de entrenamiento y el score de test ayuda a detectar si el modelo generaliza bien o si puede estar sobreajustando los datos de entrenamiento.


## 10. Probar diferentes valores de `k`

El parámetro `n_neighbors` indica cuántos vecinos utiliza KNN para clasificar una observación.

Probamos varios valores de `k` para ver cómo cambia el rendimiento del modelo.


```python
# Probar distintos valores de k
for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    score = knn.score(X_test, y_test)
    print("k:", k, "- Test score:", score)
```

### Interpretación de los valores de `k`

Al probar distintos valores de `k`, podemos observar cómo cambia el rendimiento del modelo.

El objetivo es identificar un número de vecinos que obtenga un buen resultado en test sin depender demasiado del conjunto de entrenamiento.

Un valor de `k` demasiado pequeño puede hacer que el modelo sea muy sensible a casos concretos. Un valor demasiado grande puede hacer que el modelo sea demasiado general.


## Conclusión final

En este lab hemos construido un primer modelo de Machine Learning usando `KNeighborsClassifier`.

Como la variable objetivo `Transported` es binaria (`True` / `False`), se trata de un problema de clasificación. Por ese motivo usamos un clasificador y no un regresor.

El modelo se ha entrenado con el conjunto de entrenamiento y se ha evaluado con el conjunto de test usando `.score()`, que en clasificación devuelve el accuracy.

El resultado obtenido permite hacernos una primera idea de la calidad del modelo. Además, comparar el score de entrenamiento y el score de test ayuda a comprobar si el modelo generaliza correctamente con datos no vistos.
